```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Dataset: CO2 Emissions by Country 2000-2022
# Source: Our World in Data (https://ourworldindata.org/co2-emissions)
df = pd.read_csv('/content/co2_emissions.csv')
print(f"Loaded: {len(df)} rows | Countries: {df['Country'].nunique()} | Years: {df['Year'].min()}-{df['Year'].max()}")
print(df.head())
```

```python
# Explore before building

print("Countries:", df['Country'].unique())
print("\nCO2 range:", df['CO2_Mt'].min(), "to", df['CO2_Mt'].max(), "Mt")
print("\nRegional averages (2022):")
print(df[df['Year']==2022].groupby('Region')['CO2_Mt'].mean().sort_values(ascending=False).round(1))
```

---
## Task 1 — Multi-Series Line Chart with Highlight

**What to build:** A line chart showing CO2 emissions over time for **all Asian countries** in the dataset, with one country highlighted.

**Requirements:**
- All countries shown (for context), but only **one highlighted in colour** — your choice which
- All other lines in grey (#DDDDDD), thinner
- Highlighted country **labelled directly** at the end of its line (not in a legend)
- Insight title that names the highlighted country and its story

> 💡 `df[df['Region'] == 'Asia']` to filter; use `go.Figure()` with a loop for per-country control


```python
# Task 1 — Multi-series line with highlight

# Filter Asian countries
asia_df = df[df['Region'] == 'Asia']

# Choose one country to highlight (example: India)
highlight_country = "India"

# Create base line chart
fig = px.line(
    asia_df,
    x="Year",
    y="CO2_Mt",
    color="Country",
    title="CO₂ Emissions Over Time (Asia)"
)

# Reduce opacity for all lines
for trace in fig.data:
    trace.update(opacity=0.3)

# Highlight selected country
highlight_df = asia_df[asia_df['Country'] == highlight_country]

fig.add_trace(
    go.Scatter(
        x=highlight_df["Year"],
        y=highlight_df["CO2_Mt"],
        mode="lines",
        name=highlight_country,
        line=dict(width=4),
        opacity=1
    )
)

fig.show()
```

## Task 2 — Slopegraph: Regional Change 2000 vs 2022

**What to build:** A slopegraph comparing **average regional CO2 emissions** between 2000 and 2022.

**Requirements:**
- One line per region (not per country — aggregate first)
- Colour: regions that increased = one colour; decreased = another
- Values labelled at both ends of each line
- No y-axis tick labels (the endpoint labels make them redundant)
- Insight title stating which regions moved most

> 💡 `df.groupby(['Region','Year'])['CO2_Mt'].mean().reset_index()` then filter to 2000 and 2022


```python
# Task 2 — Slopegraph: regional averages

# Filter for years 2000 and 2022
df_2000 = df[df['Year'] == 2000]
df_2022 = df[df['Year'] == 2022]

# Calculate regional averages
avg_2000 = df_2000.groupby('Region')['CO2_Mt'].mean()
avg_2022 = df_2022.groupby('Region')['CO2_Mt'].mean()

regions = avg_2000.index

# Create slopegraph
fig = go.Figure()

for region in regions:
    fig.add_trace(
        go.Scatter(
            x=[2000, 2022],
            y=[avg_2000[region], avg_2022[region]],
            mode='lines+markers+text',
            text=[region, region],
            textposition='top center',
            name=region
        )
    )

fig.update_layout(
    title="Regional CO₂ Emissions Change (2000 vs 2022)",
    xaxis=dict(title="Year", tickvals=[2000, 2022]),
    yaxis_title="Average CO₂ Emissions (Mt)",
    showlegend=False
)

fig.show()
```

The visualizations show that CO₂ emissions in Asia have increased steadily over time, with significant variation between countries. The slopegraph further highlights that Asia experienced the largest growth between 2000 and 2022, while some regions like Europe remained stable or declined.”
